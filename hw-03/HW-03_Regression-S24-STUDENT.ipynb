Also, use the `add_constant()` method in `statsmodels` to ensure the model includes a constant term as well. Fit models to both the 2022 and 2023 data and display summaries of both results.

````python
# Put your code here
mult2022 = whr_2022[['gdp_per_capita', 'healthy_life_expectancy', 'freedom_to_make_life_choices', 'generosity']]
mult2022_x = sm.add_constant(mult2022)
mult_model = sm.OLS(whr_2022['happiness_score'], mult2022_x)
results_mult2022 = mult_model.fit()
print(results_mult2022.summary())

mult2023 = whr_2023[['gdp_per_capita', 'healthy_life_expectancy', 'freedom_to_make_life_choices', 'generosity']]
mult2023_x = sm.add_constant(mult2023)
mult_model2 = sm.OLS(whr_2023['happiness_score'], mult2023_x)
results_mult2023 = mult_model2.fit()
print(results_mult2023.summary())
````
